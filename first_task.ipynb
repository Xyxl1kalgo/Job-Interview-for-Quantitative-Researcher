```python
import pandas as pd

df_order_book = pd.read_csv('order_book_snapshot.csv')
df_trades = pd.read_csv('trades.csv')

missing_values_order_book = df_order_book.isnull().sum()
columns_with_missing_order_book = missing_values_order_book[missing_values_order_book > 0]

print('OrderBook Data')
if not columns_with_missing_order_book.empty:
    print(columns_with_missing_order_book)
else:
    print('All data is complete')

missing_values_trades = df_trades.isnull().sum()
columns_with_missing_trades = missing_values_trades[missing_values_trades > 0]

print('Trades Data')
if not columns_with_missing_trades.empty:
    print(columns_with_missing_trades)
else:
    print('All data is complete')

print('df_order_book info')
print(df_order_book.info())

print('df_trades info')
print(df_trades.info())

#additional checking and work with time
df_order_book['timestamp'] = pd.to_datetime(df_order_book['timestamp'], unit='ns')
print('TimeStamp in Datatime ORDERBOOK')
print(df_order_book.head())

df_trades['timestamp'] = pd.to_datetime(df_trades['timestamp'], unit='ns')
print('TimeStamp in Datatime TRADES')
print(df_trades.head())
```
